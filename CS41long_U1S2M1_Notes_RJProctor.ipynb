**Build Data Driven Machine**

Components
1. memory
2. register
3. running loop
  * fetch, decode, run
  * logic
  * functionality
4. operational codes/operations


```
# in CPU opcodes
LOAD_NUM = 0b00000000  #0 
PRINT_NUM = 0b00000001  #1
PRINT_NAME = 0b00000010 #2
PRINT_REG = 0b00000010  #3
HALT = 0b00000100      #4
ADD = 0b00000101       #5



# in CPU memory
#ram = [0]*255  # limits to 255
ram = [
       PRINT_NAME,
       PRINT_NAME,
       PRINT_NUM,
       123,
       LOAD_NUM,
       100,
       0,
       LOAD_NUM,
       20,
       1,
       ADD,
       PRINT_REG,
       0,
       HALT
]

#in CPU program counter
pc = 0

# in CPU registers (r0-r7)
registers = [0]*8 # limits to 8-bit

# set running loop
running = True

# running loop
while running:
  # fetch
  inst = ram[pc]
  
  # decode
  if inst == PRINT_NAME:
    # execute
    print("Jeannine")
    pc +=1
  
  # decode
  elif inst == HALT:
    # execute
    running = False
    pc +=1
  
  #decode
  elif inst == LOAD_NUM:
    # execute
    # get number
    num = ram[pc+1]
    # get register index
    reg_index = ram[pc+2]
    # put number in register list at index of register 
    registers[reg_index] = num
    pc +=3
  
  #decode
  elif inst == PRINT_NUM:
    #execute
    print(ram[pc+1])
    pc +=2

 #decode
  elif inst == PRINT_REG:
    #execute
    reg_index = ram[pc+1]
    print(registers[reg_index])
    pc +=2

  #decode
  elif inst == ADD:
    #execute
    reg_index1 = ram[pc+1]
    reg_index2 = ram[pc+2]

    registers[reg_index1] += registers[reg_index2]
    pc +=3
  

  #decode
  else:
    print(f'Unknown Instruction {inst}')
    running = False

```
