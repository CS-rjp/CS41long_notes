**Bitwise Operations**

1. How would you perform an NOR operation between two numbers x and y if you didn't have an NOR operator? 



```
# NOR logic/gate

if a != b:
  return 0 # False

elif a = 1 & b = 1:
  return 0 # False

else:
  return 1 # True

```





2. How could you use the other bitwise operators to the same effect?


```
# AND gate/logic

if a != b:
  return 0 # False
 
elif a = 0 & b = 0:
  return 0 # False
 
else:
  return 1 # True
```




3. How would you perform an XOR operation between two numbers x and y if you didn't have an XOR operator? 



```
# XOR gate/logic

if a != b:
  return 1 # True
 
else:
  return 0 # False
```



4. How could you use the other bitwise operators to the same effect?


```
# NAND gate/logic
 
if a != b:
  return 1 # True
 
elif a = 0 & b = 0:
  return 1 # True
 
else:
  return 0 # False
```
