**CPU Interrupts**

*What are Interrupts?*
* commonly raised by peripherals (keyboard, mouse, etc.)
* alert to CPU that work needs to be done
* when an interrupt occurs, the current state of the proccessor is saved on the stack and the execution continues at the address of the interrupt handler
* this saves the work so the CPU can restore the state of the processor come back to it later
* after the handler completes, the processor state is poped off the stack and execution resumes (like nothing ever happened)
* most CPUs have a lookup table called an Interrupt Vector Table that is an array of pointers to functions/interrupt handlers, one pointer per interrupt (stored in RAM)


*How is an interrupt raised?*

An interrupt is a signal sent to the processor that interrupts the current process. It may be generated by a hardware device or a software program. A hardware interrupt is often created by an input device such as a mouse or keyboard. An interrupt is sent to the processor as an interrupt request, or IRQ.


*How are interrupts handled?*

The software assigns each interrupt to a handler in the interrupt table. An interrupt handler is just a routine containing a sequence of operations. Each of these may request input and output while running. Thus, an interrupt can be handled either as a thread or as a sub-process within a task or process.

An IRQ or FIQ exception causes the processor hardware to go through a standard procedure (provided the interrupts are not masked): 
1. The processor changes to a specific interrupt request mode, which reflects the interrupt being raised.

2. The previous mode's cpsr is saved into the spsr of the new interrupt request mode.

3. The pc is saved in the lr of the new interrupt request mode.

4. Interrupt/s are disabledâ€”either the IRQ or both IRQ and FIQ exceptions are disabled in the cpsr. This immediately stops another interrupt request of the same type being raised.

5. The processor branches to a specific entry in the vector table.

*Why are Interrupts useful?*

Interrupts are important because they give the user better control over the computer. Without interrupts, a user may have to wait for a given application to have a higher priority over the CPU to be ran. This ensures that the CPU will deal with the process immediately.
