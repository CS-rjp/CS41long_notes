**RAM**

rando, access memory

fast compared to hard drive

think of as a big array
with an address that works like an index 

memory bus is a collection of wires that connects the CPU to the RAM

**CPU**

while RAM stores in bytes, CPU operates on words

words are large numbers, stored sequentially in RAM that the CPU can opererate on

the exact number of bytes per word depends on architecture
*  8 bytes for a 64-bit CPU
*  4 bytes for a 32-bit CPU
*  1 byte for an 8-bit CPU

this represents how much a CPU can operate on at once





**CPU Registers**

registers store words that can be accessed at ultra-hig-speed

think of them as variables that the CPU has at disposal, simillarly to RAM except much faster (super fast) because the CPU has direct access to them

they are limited in number - usually depends on the size of CPU (8, 16, 32, 64)

they have fixed names depending on CPU (R01, R02, ..., R07)

many can CPUs only perform math operations on registers

**CPU Instructions**

stored in RAM with other data (opcodes)

are numbers - just like the other data in RAM

human-readable form of instructions are mnemonics

the CPU decides what to do based on the currently-executing instruction (opcode) it pulls out of RAM

the CPU keeps tracks of the address of currently-executing instruction in RAM and performs different actions based on the instruction found in the RAM; it does this because instructions are given in sequential order

the address of the currently-executing instruction is held in a special register called the program counter (PC); the PC is the index into the array of the currently-executing instruction

CPUs usually have a significant number of instructions (around 50-200)


**CPU Clock**

the CPU clock triggers a few billion times per second

the clock cycle is measured in:  (billions of cycles per second)
* Hz
* KHz
* MHz
* GHz

each instruction takes one or more clock cycles to execute

each time the clock cycles it executes an instruction, but some instructions take more than ome cylce to complete

in general, the faster the clock, the more instructions your CPU can execute per second


**Concurrency**

How do we get the CPU to do more than one thing at once?

each hardware component can only do one thing at a time
* duplicate the hardware components
* mutli-core CPUs
* pipelining
* timesharing (not true parallelism)



**System Bus**

How data is passed from the CPU to RAM, and from the CPU to peripherals

1. *Address Bus*
carries the adddress or peripheral ID in RAM we are iterested in

2. *Data Bus*
carries the data transmitted to/from RAM or peripherals

3. *Control Bus*
controls whether or not the CPU is talking to RAM or a pheripheral

the size or width of the bus is typicall the number of 'bits' a computer is advertised as having (8, 16, 32, 64)












**Caching**

while access to RAM is relatively slow and access to registers is fast, access to cache is the middle ground or average speed because it acts as an expiditor and speeds up access to RAM by grabbing chunks of addresses/data at once and storing them for future efficient access

cache is usually arranged in a level hierachy
* cache miss - 
you are trying to access memory that is not yet in cache
* cache hit - 
the memory you want is in the cache already

3. Why are registers necessary? 

```
Registers are a type of computer memory used to quickly accept,
store, and transfer data and instructions that are being used
immediately by the CPU. The registers used by the CPU are
often termed as Processor registers.

A processor register may hold an instruction, a storage address, or
any data (such as bit sequence or individual characters).

The computer needs processor registers for manipulating data
and a register for holding a memory address. The register
holding the memory location is used to calculate the address of
the next instruction after the execution of the current
instruction is completed.
```

4. Why not just use RAM?

```
RAM is what is called volatile storage, meaning the contents of
RAM are only there as long as power is on. Once you cut power,
the contents of the RAM is deleted and not recoverable. Using RAM as
storage is impossible if you wish to maintain permanent copies of
the documents.
```

5. Why is cache useful?

```
Caches are useful because they can increase the speed of the average
memory access, and they do so without taking up as much physical
space as the lower elements of the memory hierarchy. They ameliorate
the (performance critical) memory access time by leveraging spatial
and temporal locality.
```

6. What is a CPU word?

```
A word is a fixed-sized piece of data handled as a unit by the
instruction set or the hardware of the processor. The number of bits
in a word (the word size, word width, or word length) is an
important characteristic of any specific processor design or
computer architecture.
```

7. What is the system bus and how wide is it?

```
Bus width refers to the number of bits that can be sent to the CPU
simultaneously, and bus speed refers to the number of times a
group of bits can be sent each second. A bus cycle occurs every
time data travels from memory to the CPU.
```

8. Describe the pins that are on a CPU chip.

```
The pins on a CPU chip are used as input and outputs to the CPU. We
say that a signal is "asserted" if the CPU is requesting the
operation that is conveyed by the signal. To assert a request, the
manufacturer of the CPU can choose either the value 0 or the value 1
for that signal.
```

9. What is a CPU instruction?

```
An instruction set is a group of commands for a CPU in machine
language. All CPUs have instruction sets that enable
commands to the processor directing the CPU to switch the
relevant transistors. Some instructions are simple read,
write and move commands that direct data to different hardware.
```


**Convert and Understand Decimal, Binary and Hexadecimal**

````
# Convert Binary to Decimal in Python
def decimalToBinary(n):
  return bin(n).replace("0b","")

# driver code
if __name__ == "__main__":
  print(decimalToBinary(8))
  print(decimalToBinary(18))
  print(decimalToBinary(7))
  
````

````
# Convert Decimal to Binary in Python
def binaryToDecimal(n):
  return int(n, 2)

# driver code
if __name__ == "__main__":
  print(binaryToDecimal('100'))
  print(binaryToDecimal('101'))
  print(binaryToDecimal('1001'))
````
